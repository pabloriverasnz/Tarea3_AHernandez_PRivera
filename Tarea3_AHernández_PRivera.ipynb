### Ejercicio 1: 
1. Resolver la siguiente ecuación recursiva usando funciones como se vió en clase
$$ D_{n}=(n-1) D_{n-1}+(n-1) D_{n-2} \quad n\ge 3$$
con $D_1=0$ y $D_2 = 1$

```python
# Respuesta Pablo
import numpy as np
```

```python
def ecuacion_rec(n):
    res = np.zeros(n)
    res[0], res[1] = 0, 1
    
    def vector(i):
        nonlocal res
        res[i - 1] = (i - 1) * res[i - 2] + (i - 1) * res[i - 3]
        
    [vector(i) for i in range(3, n + 1)]
    
    return res

solec = ecuacion_rec(10)
np.set_printoptions(suppress=True)
print(solec)
```

Se llegó a la solución de este problema creando una función anidada con la ayuda de variables locales. La función vector nos va a calcular cada término de la ecuación, se agregó un nonlocal a la variable res para que se pudiera modificar fuera de la función. Después, en la función de afuera (ecuacion_rec), simplemente llamamos a la función vector y se corre un ciclo for en una compresión de listas desde 3 a n para que se vaya calculando cada valor.
